```python
# Find the student name who scored maximum scores in all (exam, quiz and homework)?

stage1 = {"$unwind" : "$scores"}
stage2 = {"$group" : {"_id" : "$_id", "totalScore" : {"$sum" : "$scores.score"}}}
stage3 = {"$sort" : {"totalScore" : -1}}
stage4 = {"$limit" : 1}
for i in col.aggregate([stage1, stage2, stage3, stage4]):
  print(i)
```

```python
# Find students who scored below average in the exam and pass mark is 40%?

stage1 = {"$unwind" : "$scores"}
stage2 = {"$match" : {"scores.type" : "exam"}}
stage3 = {"$match" : {"$expr" : {"$and" : [{"$gte" : ["$scores.score", 40]}, {"$lte" : ["$scores.score", 60]}]}}}
stage4 = {"$project" : {"_id" : 0}}

for i in col.aggregate([stage1, stage2, stage3, stage4]):
  print(i)
```

```python
# Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.

stage1 = {"$unwind" : "$scores"}
stage2 = {"$project" :
          {
              "_id" : 0,
           "name" : 1,
           "scores.type" : 1,
           "scores.score" : 1,
           "Pass/Fail" : {
               "$cond" : {
                   "if" : {"$gte" : ["$scores.score" , 40]},
                   "then" : "Pass",
                   "else" : "Fail"
               }
               
           }
          }} 

for i in col.aggregate([stage1, stage2]):
  print(i)
```

```python
# Find the total and average of the exam, quiz and homework and store them in a separate collection.

stage1 = {"$unwind" : "$scores"}
stage2 = {"$group" : {"_id" : "$_id" , "avg_score" : {"$avg" : "$scores.score"}, "total" : {"$sum" : "$scores.score"}}}
stage3 = {"$out" : "avg_and_total"}

for i in col.aggregate([stage1, stage2, stage3]):
  print(i)
```

```python
col1 = db["avg_and_total"]

for i in col1.find():
  pprint(i)
```

```python
# Create a new collection which consists of students who scored below average and above 40% in all the categories.
# score (greater than or equal to 60) and (lesser than or eaqual to 40) considered as below average score

stage1 = {"$unwind" : "$scores"}
stage2 = {"$match" : {"$expr" : {"$and" : [{"$gte" : ["$scores.score", 40]}, {"$lte" : ["$scores.score", 60]}]}}}
stage3 = {"$project" : {"_id" : 0}}
stage4 = {"$out" : "below_avg_students_list"}

for i in col.aggregate([stage1, stage2, stage3, stage4]):
  print(i)
```

```python
col2 = db["below_avg_students_list"]

for i in col2.find({}, {"_id" : 0}):
  print(i)
```

```python
# Create a new collection which consists of students who scored below the fail mark in all the categories

stage1 = {"$unwind" : "$scores"}
stage2 = {"$match" : {"$expr" : {"$lt" : ["$scores.score", 40]}}}
stage3 = {"$sort" : {"scores.score" : -1}}
stage4 = {"$project" : {"_id" : 0}}
stage5 = {"$out" : "Failed_students_list"}

for i in col.aggregate([stage1, stage2, stage3, stage4, stage5]):
  print(i)
```

```python
col3 = db["Failed_students_list"]

for i in col3.find({}, {"_id" : 0}):
  print(i)
```

```python
# Create a new collection which consists of students who scored above pass mark in all the categories.

stage1 = {"$unwind" : "$scores"}
stage2 = {"$match" : {"$expr" : {"$gte" : ["$scores.score", 40]}}}
stage3 = {"$sort" : {"scores.score" : -1}}
stage4 = {"$project" : {"_id" : 0}}
stage5 = {"$out" : "Passed_students_list"}

for i in col.aggregate([stage1, stage2, stage3, stage4, stage5]):
  print(i)
```

```python
col4 = db["Passed_students_list"]

for i in col4.find({}, {"_id" : 0}):
  print(i)
```
