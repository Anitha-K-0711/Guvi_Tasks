```python
db = cluster.sample_airbnb
```

```python
import pandas as pd
```

```python
collection = db.listingsAndReviews
```

```python
l = list((collection.find()))
```

```python
df = pd.DataFrame(l)
```

```python
df.head()
```

```python
df.columns
```

```python
df.describe()
```

```python
df[df["bedrooms"]>=2]
```

```python
df[df["bedrooms"] == 1]
```

```python
df.groupby("property_type")["reviews_per_month"].mean().reset_index()
```

```python
# to know the average weekly price for each property type
df["weekly_price"] = df["weekly_price"].astype(str).astype(float)
df.groupby("property_type")["weekly_price"].mean().reset_index()
```

```python
# to know the average monthly price for each property type
df["monthly_price"] = df["monthly_price"].astype(str).astype(float)
df.groupby("property_type")["monthly_price"].mean().reset_index()
```

```python
df.groupby("property_type")["weekly_price","monthly_price"].mean().reset_index()
```

```python
df["price_variety_for_monthly_price"] = pd.cut(df["monthly_price"], bins = 3, labels = ["cheap", "medium price", "costlier"])
```

```python
df.groupby("price_variety_for_monthly_price")["accommodates"].mean().reset_index()
```

```python
df[["bedrooms", "property_type"]]
```

```python
df.dtypes
```

```python
# To know the number of people accommodating in each property type
df.groupby("property_type")["accommodates"].sum()
```

```python
df.groupby("property_type")[df["accommodates"] > 200].sum() 
```

```python
# To get all the apartments from property type 
df[df["property_type"] == "Apartment"]
```

```python
# to get the list of apartments having 2 bedrooms
df[df["property_type"] == "Apartment"][df["bedrooms"] == 2]
```

```python
# to find 2 bedroom apartments in New York
df[df["property_type"] == "Apartment"][df["bedrooms"] == 2]df["address"].str.contains('NY').sum()
```

```python
df["price"] = df["price"].astype(str).astype(float)
df["security_deposit"] = df["security_deposit"].astype(str).astype(float)
df["cleaning_fee"] = df["cleaning_fee"].astype(str).astype(float)
df["extra_people"] = df["extra_people"].astype(str).astype(float)
```

```python
# To know the average  of price, cleaning fee, security deposit, extra people fee of room types
df.groupby("room_type")["price", "security_deposit", "cleaning_fee", "extra_people"].mean().reset_index()
```

```python
# To know the average  of price, cleaning fee, security deposit, extra people fee of property types
df.groupby("property_type")["price", "security_deposit", "cleaning_fee", "extra_people"].mean().reset_index()
```

```python
df["bathrooms"] = df["bathrooms"].astype(str).astype(float)
```

```python
df.groupby("property_type")["bedrooms", "bathrooms"].mean().reset_index()
# How to get this average value as a whole number?
```
